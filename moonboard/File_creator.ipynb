```python
f = open('problems2017.txt')
l = f.readline()
l = f.readline()
d = []
while l != ']':
    d.append(eval(l))
    l = f.readline()
t = type(d[0])
for i,dic in enumerate(d):
    if type(dic) == t:
        d[i] = d[i][0]
        
```

```python
def moveToArray(lst):
    board = np.zeros((3,11,18))
    for string in lst:
        splt = string.split(" ")
        x = ord(splt[0][0])-65
        y = int(splt[0][1:]) - 1
        if splt[1][0] == 'T':
            board[0][x][y] = 1
        elif splt[2][0] == 'T':
            board[2][x][y] = 1
        else:
            board[1][x][y] = 1
    return board
for i,dic in enumerate(d):
    d[i]['moves'] = moveToArray(d[i]['moves'])
```
